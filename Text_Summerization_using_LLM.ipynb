```python
from datasets import load_dataset
from transformers import pipeline
```

```python
xsum_dataset = load_dataset(
    "xsum",
    version="1.2.0",
    cache_dir='/Documents/Huggin_Face/data'
)  # Note: We specify cache_dir to use predownloaded data.
xsum_dataset
```

```python
xsum_sample = xsum_dataset["train"].select(range(10))

display(xsum_sample.to_pandas())
```

## Create Summerization Pipeline

```python
summerizer = pipeline(
    task = "summarization",
    model = "t5-small",
    min_length = 20,
    max_length = 40,
    truncation = True,
    model_kwargs = {"cache_dir":'/Documents/Huggn_Face/'},
)
```

```python
summerizer(xsum_sample["document"][0])
```

```python
input_text = input("Enter the text you want to summarize: ")

# Generate the summary
summary = summerizer(input_text, max_length=150, min_length=30, do_sample=False)[0]['summary_text']

bullet_points = summary.split(". ")

for point in bullet_points:

    print(f"- {point}")

# Print the generated summary
print("Summary:", summary)
```
